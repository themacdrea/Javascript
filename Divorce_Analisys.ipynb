<h1 style = "font-family:'Arial Narrow';font-size:2rem;color:black;">Cleaning and Structuring Data for Effective Analysis</h1>

```python
# Import dependencies
import pandas as pd
from datetime import datetime
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
# Set the option to display all columns
pd.set_option('display.max_columns', None)

# Read the DataFrame from the original file
divorce_df = pd.read_csv('Resources/divorces_2000-2015_translated.csv')
divorce_df.head()
```

```python
# Remove unwanted columns
remove = ['Place_of_birth_partner_man',
'Birth_municipality_of_partner_man',
'Birth_federal_partner_man',
'Birth_country_partner_man', 
'Residence_municipality_partner_man',
'Residence_federal_partner_man',
'Residence_country_partner_man',  
'Place_of_residence_partner_man',
'DOB_registration_date_partner_woman',
'Place_of_birth_partner_woman',
'Birth_municipality_of_partner_woman',
'Birth_federal_partner_woman',
'Birth_country_partner_woman',
'Place_of_residence_partner_woman',
'Residence_municipality_partner_woman',
'Residence_federal_partner_woman',
'Residence_country_partner_woman', 
'Marriage_certificate_place',
'Marriage_certificate_municipality',
'Marriage_certificate_federal',
'Marriage_duration_months',
'Custody']

divorce_df = divorce_df.drop(columns=remove)
divorce_df
```

```python
# Drop any rows in the DataFrame that contain missing values (NaN)
divorce_df.dropna(inplace=True)

# Reset the index after removing rows
divorce_df.reset_index(drop=True, inplace=True)
divorce_df
```

```python
# Get the divorce_df columns.
divorce_df.columns.to_list()
```

```python
# Iterate over each column in the columns list. 
columns = ['Divorce_date',
 'Type_of_divorce',
 'Nationality_partner_man',
 'DOB_partner_man',
 'Age_partner_man',
 'Monthly_income_partner_man_peso',
 'Occupation_partner_man',
 'Nationality_partner_woman',
 'DOB_partner_woman',
 'Age_partner_woman',
 'Occupation_partner_woman',
 'Monthly_income_partner_woman_peso',
 'Date_of_marriage',
 'Level_of_education_partner_man',
 'Employment_status_partner_man',
 'Level_of_education_partner_woman',
 'Employment_status_partner_woman',
 'Marriage_duration',
 'Num_Children']

# Calculate the total number of null values for each column, and print the result to see if there is more null values
for col in columns:
    total_nulls = divorce_df[col].isnull().sum()
    print("Total null values in column '{}' are: {}".format(col, total_nulls))
```

<h1 style = "font-family:'Arial Narrow';font-size:2rem;color:black;">Formatting Dates: Achieving Uniformity and Clarity in Data Representation</h1>

```python
# Convert the dates to a consistent format
divorce_df['Divorce_date'] = pd.to_datetime(divorce_df['Divorce_date'], format='%d/%m/%y')
```

```python
# Add the century to two-digit years 'DOB_partner_man'
divorce_df['DOB_partner_man'] = divorce_df['DOB_partner_man'].str.replace(r'^(\d{1,2})/(\d{1,2})/(\d{2})$', r'\1/\2/19\3')

# Convert the dates to datetime objects
divorce_df['DOB_partner_man'] = pd.to_datetime(divorce_df['DOB_partner_man'], format='%d/%m/%Y')
```

```python
# Add the century to two-digit years 'DOB_partner_woman'
divorce_df['DOB_partner_woman'] = divorce_df['DOB_partner_woman'].str.replace(r'^(\d{1,2})/(\d{1,2})/(\d{2})$', r'\1/\2/19\3')

# Convert the dates to datetime objects
divorce_df['DOB_partner_woman'] = pd.to_datetime(divorce_df['DOB_partner_woman'], format='%d/%m/%Y')
```

```python
# Convert the dates to a four-digit year format 'Date_of_marriage'
divorce_df['Date_of_marriage'] = pd.to_datetime(divorce_df['Date_of_marriage'], format='%d/%m/%y').dt.strftime('%Y-%m-%d')
```

```python
# Create a new column with id number 
divorce_df.insert(0, 'Id', range(1, len(divorce_df)+1))
divorce_df.head(15)
```

```python
#Rename columns
column_names = {
    'Divorce_date': 'Divorce Date',
    'Type_of_divorce': 'Divorce Type',
    'Nationality_partner_man': 'Nationality Male',
    'DOB_partner_man': 'DOB Male',
    'Age_partner_man': 'Age Male',
    'Monthly_income_partner_man_peso': 'Income Male',
    'Occupation_partner_man': 'Occupation Male',
    'Nationality_partner_woman': 'Nationality Female',
    'DOB_partner_woman': 'DOB Female',
    'Age_partner_woman': 'Age Female',
    'Occupation_partner_woman': 'Occupation Female',
    'Monthly_income_partner_woman_peso': 'Income Female',
    'Date_of_marriage': 'Marriage Date',
    'Level_of_education_partner_man': 'Education Male',
    'Employment_status_partner_man': 'Employment Male',
    'Level_of_education_partner_woman': 'Education Female',
    'Employment_status_partner_woman': 'Employment Female',
    'Marriage_duration': 'Marriage Year',
    'Num_Children': 'Number Children',
    }

divorce_df = divorce_df.rename(columns=column_names)
divorce_df.head()
```

```python
# Get the divorce columns.
divorce_df.columns.to_list()
```

```python
# Select columns in a list to include in male_df

selected_columns = ['Id', 'Divorce Date', 'Divorce Type', 'Nationality Male', 'DOB Male',
       'Age Male', 'Income Male', 'Occupation Male',
       'Marriage Date', 'Education Male', 'Employment Male', 'Marriage Year',
                    'Number Children']

# Create a new data frame with only male info
male_df = divorce_df[selected_columns].copy()
male_df.head()
```

```python
# Select columns in a list to include in female_df

selected_columns = ['Id', 'Divorce Date', 'Divorce Type', 'Nationality Female',
       'DOB Female', 'Age Female', 'Occupation Female', 'Income Female',
       'Marriage Date', 'Education Female', 'Employment Female', 'Marriage Year',
                    'Number Children']

# Create a new data frame with only female info
female_df = divorce_df[selected_columns].copy()
female_df.head()
```

```python
# Define a function to assign zodiac sings to DOB

def get_zodiac_sign(date):
    day = date.day
    month = date.month

    if (month == 1 and day >= 20) or (month == 2 and day <= 18):
        return "Aquarius"
    elif (month == 2 and day >= 19) or (month == 3 and day <= 20):
        return "Pisces"
    elif (month == 3 and day >= 21) or (month == 4 and day <= 19):
        return "Aries"
    elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
        return "Taurus"
    elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
        return "Gemini"
    elif (month == 6 and day >= 21) or (month == 7 and day <= 22):
        return "Cancer"
    elif (month == 7 and day >= 23) or (month == 8 and day <= 22):
        return "Leo"
    elif (month == 8 and day >= 23) or (month == 9 and day <= 22):
        return "Virgo"
    elif (month == 9 and day >= 23) or (month == 10 and day <= 22):
        return "Libra"
    elif (month == 10 and day >= 23) or (month == 11 and day <= 21):
        return "Scorpio"
    elif (month == 11 and day >= 22) or (month == 12 and day <= 21):
        return "Sagittarius"
    else:
        return "Capricorn"

# Apply the get_zodiac_sign function to the 'DOB Male' column
male_df['Zodiac_Sign'] = pd.to_datetime(male_df['DOB Male']).apply(get_zodiac_sign)

# Display the updated DataFrame
male_df
```
